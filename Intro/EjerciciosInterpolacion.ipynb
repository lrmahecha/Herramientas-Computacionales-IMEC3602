```python
from scipy import interpolate
import matplotlib.pyplot as plt
import numpy as np
import math

x= [0.0,0.5,1.0,1.5]
y= [1.0,2.0,1.5,1.0]

inter = interpolate.splrep(x,y)
```

```python
y_punto=interpolate.splev(2.5, inter)
```

```python
plt.plot(x,y,'*')
valores_x= np.linspace(0,3,100)
valores_y= interpolate.splev(valores_x,inter)
plt.plot(valores_x,valores_y)
plt.plot(2.5,y_punto,'x')
```

```python
x_temp = [100,120,135,150]
y_vapor = [1.0,1.96,3.09,4.70]

poly = interpolate.lagrange (x_temp,y_vapor)

valores_temp=[112,114,120,136,148]
valores_vap= poly(valores_temp)
```

```python
plt.plot(x_temp,y_vapor,'*')
valores_curvatemp= np.linspace(90,160,100)
valores_curvavap= poly(valores_curvatemp)
plt.plot(valores_curvatemp,valores_curvavap)
plt.xlabel("Temperatura(°C)")
plt.ylabel("Presión de vapor(atm)")
```

```python
grad_vel=np.array([50,70,90,110,130])
esf_cort=np.array([6.01,7.48,8.59,9.19,10.21])
```

```python
log_gradvel=[math.log(y)for y in grad_vel]
log_esfcort=[math.log(y)for y in esf_cort]
```

```python
fit=np.polyfit(log_gradvel,log_esfcort,1)
fit_fn = np.poly1d(fit)
```

```python
plt.plot(log_gradvel,log_esfcort,'*')
valores_gradvel= np.linspace(3.5,5,100)
valores_esfcort= fit_fn(valores_gradvel)
plt.plot(valores_gradvel,valores_esfcort)
plt.xlabel("$\gamma$")
plt.ylabel("$\\tau$")
```

```python
A=math.exp(fit_fn[0])
M=fit_fn[1]
```

```python
def polinomial(x):
    
    y=A*(x**M)
    
    return y
```

```python
x_poli=np.linspace(40,140,100)
y_poli=polinomial(x_poli)
plt.plot(x_poli,y_poli)
plt.plot(grad_vel,esf_cort,'*')
plt.xlabel("$\gamma$")
plt.ylabel("$\\tau$")
```
