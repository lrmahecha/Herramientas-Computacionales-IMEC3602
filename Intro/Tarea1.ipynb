```python
from numpy import pi
from random import random
import matplotlib.pyplot as plt

def Pi(total):
    
    in_c = 0
    
    for i in range(total):
        
        x = random()
        y = random()
        d = (x ** 2 + y ** 2-1)
        
        if d < 0:
            in_c += 1
            
    return in_c / total

valor=4*Pi(100000)
print("El valor de pi es")
print(valor)

valores_pi=[]

n=[10,10**2,10**3,10**4,10**5,10**6,10**7,10**8]
for i in n:
    
    puntos= Pi(i)
    valores_pi.append(puntos-pi/4)
    
print("grafica convergencia")
    
plt.semilogx(n,valores_pi)
plt.show()
```
